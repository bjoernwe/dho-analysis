## Typed DataFrame PoC

Reads `data/messages.jsonl` into a `DataFrame<Message>` — all column access is via property references, no strings.

````kotlin
import data.Message
import data.readMessages
import org.jetbrains.kotlinx.dataframe.api.*

val df = readMessages("../../data/messages.jsonl")
println("Total rows : ${df.rowsCount()}")
println("Columns    : ${df.columnNames()}")
````

````kotlin
val practiceLogs = df.filter { it.category == "PracticeLogs" }
println("PracticeLogs: ${practiceLogs.rowsCount()} rows, ${practiceLogs[Message::author].distinct().size()} unique authors")
````

````kotlin
practiceLogs.groupBy(Message::author).count().sortByDesc("count").head(10)
````

````kotlin
practiceLogs.filter { it.author == "Linda ”Polly Ester” Ö" }.filter { it.threadAuthor == "Linda ”Polly Ester” Ö" }.count()
````
